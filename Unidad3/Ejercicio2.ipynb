#Problema 2: Distribución Exponencial

Contexto: Una estación de carga rápida para vehículos eléctricos quiere simular el tiempo entre la llegada de los vehículos para prever la cantidad de puntos de carga necesarios durante el día. La llegada de los vehículos sigue una distribución exponencial con una tasa de 1 llegada cada 15 minutos, en promedio. Objetivo: Generar tiempos entre llegadas de vehículos eléctricos durante un día de operación (12 horas).

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro de la distribución exponencial (λ = 1 / tiempo promedio entre llegadas)
lambda_llegadas = 1 / 15  # 1 llegada cada 15 minutos

# Duración de la simulación (12 horas = 720 minutos)
horas_operacion = 12
minutos_totales = horas_operacion * 60

# Función para generar tiempos entre llegadas usando transformada inversa
def generar_tiempo_llegada(lambda_llegadas):
    u = np.random.uniform(0, 1)  # Número aleatorio uniforme
    tiempo_llegada = -np.log(1 - u) / lambda_llegadas  # Fórmula exponencial inversa
    return tiempo_llegada

# Simular llegadas hasta cubrir las 12 horas
tiempos_llegadas = []
tiempo_acumulado = 0

while tiempo_acumulado < minutos_totales:
    tiempo_entre_llegadas = generar_tiempo_llegada(lambda_llegadas)
    tiempo_acumulado += tiempo_entre_llegadas
    if tiempo_acumulado < minutos_totales:  # Evitar exceder el tiempo de simulación
        tiempos_llegadas.append(tiempo_entre_llegadas)

# Mostrar resultados
print("Tiempos entre llegadas simulados (minutos):")
print(tiempos_llegadas[:10], "...")  # Mostrar solo los primeros 10 para brevedad
print(f"\nTotal de llegadas en {horas_operacion} horas: {len(tiempos_llegadas)}")

# Graficar histograma de tiempos entre llegadas
plt.hist(tiempos_llegadas, bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribución de Tiempos entre Llegadas de Vehículos (Exponencial)')
plt.xlabel('Tiempo entre llegadas (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```
